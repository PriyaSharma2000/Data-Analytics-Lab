```python
import pandas as pd
df1 = pd.read_csv('AWCustomers.csv')
print(df1.head())
print(df1.columns)
```

```python
selected_cols = ['Education', 'Occupation', 'Gender', 'MaritalStatus', 'HomeOwnerFlag', 
                 'NumberCarsOwned', 'NumberChildrenAtHome', 'TotalChildren', 'YearlyIncome']

df_selected = df1[selected_cols]
df_selected.head()
```

```python
attribute_types = {
    'YearlyIncome': ('Continuous', 'Ratio'),
    'Gender': ('Discrete', 'Nominal'),
    'MaritalStatus': ('Discrete', 'Nominal'),
    'TotalChildren': ('Discrete', 'Ratio'),
    'NumberChildrenAtHome': ('Discrete', 'Ratio'),
    'Education': ('Discrete', 'Ordinal'),
    'Occupation': ('Discrete', 'Nominal'),
    'HomeOwnerFlag': ('Discrete', 'Nominal (Binary)'),
    'NumberCarsOwned': ('Discrete', 'Ratio'),
    'Age': ('Continuous', 'Ratio'),
    'CommuteDistance': ('Discrete', 'Ordinal'),
    'Purchased Bike': ('Discrete', 'Nominal (Binary)')
}

df_types = pd.DataFrame([
    {'Attribute': col,
     'Pandas dtype': df_selected[col].dtype,
     'Data Type': attribute_types[col][0],
     'Measurement Scale': attribute_types[col][1]}
    for col in selected_cols
])

print(df_types)
```

```python
df2 = pd.read_csv("AWSales.csv") 
print(df2.columns);
df = pd.merge(df1,df2,on="CustomerID")
buyers = df[df["BikeBuyer"] == 1]["YearlyIncome"]
non_buyers = df[df["BikeBuyer"] == 0]["YearlyIncome"]
print("\nAverage income of buyers:", buyers.mean())
print("Average income of non-buyers:", non_buyers.mean())
```
