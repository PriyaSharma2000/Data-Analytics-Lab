```python
df = pd.read_csv("employees.csv")
rows,cols = df.shape
print("Number of rows:",rows)
print("Number of columns:",cols)
```

```python
print(df.info())
```

```python
print(df.describe(include='all'))
```

```python
print("First 5 rows:\n",df.head)
print("Last 3 rows:\n",df.tail(3))
```

```python
avg_salary = df['Salary'].mean()
print("Average Salary:",avg_salary)
total_bonus = df['Bonus'].sum()
print("Total Bonus:", total_bonus)
youngest_age = df['Age'].min()
print("Youngest Employee Age:",youngest_age)
highest_rating = df['Rating'].max()
print("Highest Performance rating:", highest_rating)
```

```python
sort_values = df.sort_values(by = 'Salary',ascending=False)
print("Sort values by salary:\n",sort_values)
```

```python
def categorize(rating):
    if rating>=4.5:
        return "Excellent"
    elif rating >= 4.0:
        return "Good"
    else:
        return "Average"
df["Category"] = df["Rating"].apply(categorize)
print(df)
```

```python
print(df.isnull())
```

```python
df.rename(columns = {"Employee_ID":"ID"}, inplace = True)
print(df)
```

```python
emp_experience = df[df['Years_of_Experience'] > 5]
print("Employees with more than 5 years of experience:\n",emp_experience)
it_emp = df[df['Department'] == 'IT']
print("Employees in IT department:\n", it_emp)
```

```python
df["Tax"] = df["Salary"]*0.10
print(df)
```

```python
df.to_csv("employees_modified.csv", index=False)
```
